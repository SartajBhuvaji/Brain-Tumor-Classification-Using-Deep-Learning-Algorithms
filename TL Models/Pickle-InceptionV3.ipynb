```
TEST_DIR = '/content/gdrive/My Drive/Data/TestingData' # test data folder
TRAIN_DIR = '/content/gdrive/My Drive/Data/TrainingData' # train data folder
IMG_SIZE = 299 # image size
CATEGORIES = ["glioma","meningioma","no_tumor","pituitary"]
```

```
testing_data = []

def create_testing_data():
    for category in CATEGORIES: 

        path = os.path.join(TEST_DIR,category)  # create path 
        class_num = CATEGORIES.index(category)  # get the classification  

        for img in tqdm(os.listdir(path)): 
          # iterate over each image per dogs and cats 
          img_array = cv2.imread(os.path.join(path,img) ,cv2.IMREAD_COLOR)  # convert to array
          new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))  # resize to normalize data size
          testing_data.append([new_array, class_num])  # add this to our training_data
    
    random.shuffle(testing_data)
    #np.save('test_data.npy', testing_data)        

create_testing_data()
#np.save('test_data.npy', testing_data) 
print(len(testing_data))

print("test")
print()
X_test = np.array([i[0] for i in testing_data]).reshape(-1,IMG_SIZE,IMG_SIZE,3)
Y_test = [i[1] for i in testing_data]

import pickle

pickle_out = open("X_test_inc.pickle","wb")
pickle.dump(X_test, pickle_out)
pickle_out.close()

pickle_out = open("Y_test_inc.pickle","wb")
pickle.dump(Y_test, pickle_out)
pickle_out.close()
```
